```python
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Logistic Regression Model
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)

# Save model & scaler
with open('logistic_model.pkl', 'wb') as model_file:
    pickle.dump(log_reg, model_file)

with open('scaler.pkl', 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)
```

```python
import streamlit as st
import numpy as np
import pickle

# Load trained model & scale
with open('logistic_model.pkl', 'rb') as model_file:
    model = pickle.load(model_file)

with open('scaler.pkl', 'rb') as scaler_file:
    scaler = pickle.load(scaler_file)

# Streamlit UI
st.title("Personal Fitness Tracker - Calories Burned Predictor")

st.write("Enter your workout details below to estimate if you've burned high calories or not.")

# User Inputs
heart_rate = st.number_input("Enter Heart Rate (BPM)", min_value=50, max_value=200, step=1)
workout_duration = st.number_input("Enter Workout Duration (minutes)", min_value=1, max_value=180, step=1)
age = st.number_input("Enter Age", min_value=10, max_value=80, step=1)
weight = st.number_input("Enter Weight (kg)", min_value=30, max_value=150, step=1)

# Predict Button
if st.button("Predict"):
    # Prepare input data
    features = np.array([[heart_rate, workout_duration, age, weight]])
    features_scaled = scaler.transform(features)  # Scale input data

    # Make prediction
    prediction = model.predict(features_scaled)[0]

    # Display result
    if prediction == 1:
        st.success("High Calories Burned!")
    else:
        st.warning("Low Calories Burned.Try increasing workout intensity!")
```
